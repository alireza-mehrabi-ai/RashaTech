## Cropping


```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 

img = cv2.imread('./images/MachineVision1.png')
plt.imshow(img[...,::-1])
```

```python
print('Height = '  , img.shape[0] )
print('Width  = '  , img.shape[1] )
```

```python
croped = img[100:300 ,400:750 , :]

plt.imshow(croped[...,::-1])
```
