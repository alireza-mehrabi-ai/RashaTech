## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma"> عملیات ریاضی در OpenCV</div>
<div style="direction:rtl;text-align:right;font-family:Tahoma">
اگر از توابع 
    <b>cv2.add</b>
و 
    <b>cv2.subtract</b>
    استفاده کنیم این مشکل را مشاهده نخواهیم کرد و مقادیر کمتر از 0 همان 0 و مقادیر بیش از 255 همان 255 باقی خواهد ماند.
</div>


```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 

img = cv2.imread('./images/p.jpg')
A = np.ones_like(img , dtype='uint8') * 100

add_img = cv2.add(img , A)
subtract_img = cv2.subtract(img , A)


plt.figure(figsize=[15,10])
plt.subplot(131);plt.imshow(img[...,::-1]);plt.title("Image")
plt.subplot(132);plt.imshow(add_img[...,::-1]);plt.title("Added")
plt.subplot(133);plt.imshow(subtract_img[...,::-1]);plt.title("Subtract");
```

```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 

img1 = cv2.imread('./images/MachineVision1.png')[:400,:900,:]
img2 = cv2.imread('./images/MachineVision2.jpg')[:400,:900,:]

add_img = cv2.add(img1 ,img2)
subtract_img = cv2.subtract(img1 ,img2)


plt.figure(figsize=[15,10])
plt.subplot(221);plt.imshow(img1[...,::-1]);plt.title("img1")
plt.subplot(222);plt.imshow(img2[...,::-1]);plt.title("img2")
plt.subplot(223);plt.imshow(add_img[...,::-1]);plt.title("Add")
plt.subplot(224);plt.imshow(subtract_img[...,::-1]);plt.title("Subtract");
```
