## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">تبدیل تصاویر به سطح خاکستری یا Grayscaling </div>


### Solution1


```python
import matplotlib.pyplot as plt
import cv2 

image = cv2.imread('./images/MachineVision2.jpg')
cv2.imshow('Original' , image)
cv2.waitKey(0)

# We use cvtColor, to convert to grayscale
gray_img = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
cv2.imshow('Gray_Image',gray_img)
cv2.waitKey(0)

cv2.destroyAllWindows()


plt.figure(figsize=[10,8])
plt.subplot(121);plt.title('Original');plt.imshow(image[...,::-1])
plt.subplot(122);plt.title('Gray_Image');plt.imshow(gray_img ,cmap='gray')
```
