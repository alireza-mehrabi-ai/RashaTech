## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">ترسیم تصاویر و اشکال در OpenCV</div>

```python
import cv2  
import matplotlib.pyplot as plt 
import numpy as np 
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">ترسیم خط</div>
cv2.line(image, starting cordinates, ending cordinates, color, thickness)

```python
image = np.zeros((512,512,3),'uint8')
cv2.line(image ,(100,100),(400,400),(255,0,0),5)
cv2.imshow('Blue Line',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(image[...,::-1]);plt.title("Blue Line");
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">ترسیم مستطیل</div>

cv2.rectangle(image, starting vertex, opposite vertex, color, thickness)

```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 

img = np.zeros((512,512,3),np.uint8)
img2 = np.zeros((512,512,3),np.uint8)
cv2.rectangle(img ,(100,100),(400,400),(0,255,0),5)
cv2.rectangle(img2 ,(100,100),(400,400),(0,255,0),-5)
plt.subplot(121);plt.imshow(img[...,::-1]);plt.title('Rectangle1')
plt.subplot(122);plt.imshow(img2[...,::-1]);plt.title('Rectangle2');
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">ترسیم دایره</div>

cv2.cirlce(image, center, radius, color, thickness)
<div style="direction:rtl;text-align:right;font-family:Tahoma">

```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 

img = np.zeros((512,512,3),np.uint8)
cv2.circle(img ,(256,256),120,(100,0,255),10)
plt.imshow(img[...,::-1]),plt.title("Circle")
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">ترسیم بیضی</div>

```python
img = np.zeros((512,512,3),np.uint8)
cv2.ellipse(img ,(256,256),(100,150),120,0,360,(250,250,0),-5)
plt.imshow(img[...,::-1]);plt.title("Ellipse");
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">ترسیم چند ضلعی</div>


```python

img = np.zeros((512,512,3),'uint8')
points = np.array([[100,75],[100,156],[200,254],[100,500],[40,60]])
points = points.reshape((-1,1,2))

cv2.polylines(img ,[points] ,True ,(0,255,255),5)
plt.imshow(img[...,::-1]);plt.title("Polyline");
```
