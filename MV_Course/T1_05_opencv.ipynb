```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./images/transparency.png',cv2.IMREAD_UNCHANGED)
img.shape
```

```python
bgr_img = img[:,:,:3]
alpha_channel = img[:,:,-1]
rgb_img = cv2.cvtColor(bgr_img , cv2.COLOR_BGR2RGB)

plt.imshow(rgb_img );plt.title('RGB image')
```

```python
plt.imshow(alpha_channel , cmap='gray');plt.title('ALpha Channel');
```

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 


img = cv2.imread('./images/alpha.png',cv2.IMREAD_UNCHANGED)
img .shape
```

```python
bgr_img = img[...,:3]
alpha_img = img[...,-1]
rgb_img = cv2.cvtColor(bgr_img ,cv2.COLOR_BGR2RGB)

plt.figure(figsize=[12,8])
plt.subplot(121);plt.imshow(rgb_img);plt.title('RGB')
plt.subplot(122);plt.imshow(alpha_img ,cmap='gray');plt.title('Alpha')
```


<div style="direction:rtl;text-align:right;font-family:Tahoma">
مقدار $\alpha$ مقداری بین 0 تا 255 است که تقسیم بر 255 خواهد شد و فرمول زیر را خواهیم داشت: 
<br/>
پس خواهیم داشت: $0 \leq \alpha \leq 1$</div>

$I = \alpha F + (1 - \alpha) B$

<div style="direction:rtl;text-align:right;font-family:Tahoma">
طبق فرمول بالا
    <br/>
    
 - وقتی $\alpha = 0$, خروجی تصویر پس زمینه است.
 - وقتی $\alpha = 1$, خروجی تصویر شما یا زمینه است
 - وقتی $0 < \alpha < 1$ 
    یک ترکیب از تصویر و تصویر پس زمینه است
